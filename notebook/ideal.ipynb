```python
data_path="/home/redha/Documents/projects/NLP/datamining project/Soil-Fertility/data/Dataset1.csv" 
import pandas as pd
sklearn_data=pd.read_csv(data_path)
def handling(val):
    try : 
        return float(val)
    except :
        return None

sklearn_data['P']=sklearn_data['P'].apply(handling)
import numpy as np

sklearn_data.dropna(inplace=True)

sklearn_data.drop_duplicates(inplace=True)

sklearn_data.describe()
```

```python
import numpy as np
import pandas as pd

# Function to replace outliers with the mean
def replace_outliers_with_mean(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Calculate mean without outliers
    mean = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)][column].mean()

    # Replace outliers
    df[column] = np.where((df[column] < lower_bound) | (df[column] > upper_bound), mean, df[column])

# Apply the function to each column
for col in sklearn_data.columns:
    replace_outliers_with_mean(sklearn_data, col)

sklearn_data.describe()
```
